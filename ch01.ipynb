# Chapter 1: Entendendo grandes modelos de linguagem

Um LLM é uma rede neural projetada para entender, gerar e responder a texto semelhante ao humano.

O “grande” em “modelo de linguagem grande” refere-se tanto ao tamanho do modelo em termos de parâmetros quanto ao imenso dataset no qual ele é treinado.

Antes do advento dos LLMs, métodos tradicionais se destacavam em tarefas de categorização, como classificação de spam de e-mail e reconhecimento de padrões simples que podiam ser capturados com regras feitas à mão ou modelos mais simples. No entanto, eles geralmente apresentavam um desempenho inferior em tarefas linguísticas.

Por exemplo, gerações anteriores de modelos de linguagem não conseguiam escrever um e-mail a partir de uma lista de palavras-chave — uma tarefa que é trivial para os LLMs contemporâneos

Quando dizemos que os modelos de linguagem "entendem", queremos dizer que eles podem processar e gerar texto de maneiras que parecem coerentes e relevantes, não que possuam consciência ou compreensão semelhantes às humanas.

Os LLMs são treinados em grandes quantidades de dados textuais. Esse treinamento em larga escala permite que os LLMs capturem informações contextuais mais profundas e sutilezas da linguagem humana.

Os LLMs melhoraram significativamente o desempenho em uma ampla gama de tarefas de PNL, incluindo tradução de texto, análise de sentimentos, respostas a perguntas e muito mais.

Os modelos de PNL anteriores eram tipicamente projetados para tarefas específicas, os LLMs demonstram uma proficiência mais ampla em uma ampla gama de tarefas de PNL.

A previsão da próxima palavra é sensata porque aproveita a natureza sequencial inerente da linguagem para treinar modelos na compreensão de contexto, estrutura e relacionamentos dentro do texto.

Os LLMs utilizam uma arquitetura chamada de transformador, que permite que eles prestem atenção seletiva a diferentes partes da entrada ao fazer previsões

image.png

O machine learning envolve o desenvolvimento de algoritmos que podem aprender e fazer previsões com base em dados sem serem explicitamente programados.

Em vez de escrever manualmente regras para identificar e-mails de spam, um algoritmo de machine learning é alimentado com exemplos de e-mails rotulados como spam e e-mails legítimos.

O deep learning é um subconjunto do machine learning que se concentra na utilização de redes neurais com três ou mais camadas.

Em contraste com o deep learning , o machine learning tradicional requer extração manual de características. Isso significa que especialistas humanos precisam identificar e selecionar as características mais relevantes para o modelo.
