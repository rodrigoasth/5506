# Chapter 1: Entendendo grandes modelos de linguagem

Um LLM é uma rede neural projetada para entender, gerar e responder a texto semelhante ao humano.

O “grande” em “modelo de linguagem grande” refere-se tanto ao tamanho do modelo em termos de parâmetros quanto ao imenso dataset no qual ele é treinado.

Antes do advento dos LLMs, métodos tradicionais se destacavam em tarefas de categorização, como classificação de spam de e-mail e reconhecimento de padrões simples que podiam ser capturados com regras feitas à mão ou modelos mais simples. No entanto, eles geralmente apresentavam um desempenho inferior em tarefas linguísticas.

Por exemplo, gerações anteriores de modelos de linguagem não conseguiam escrever um e-mail a partir de uma lista de palavras-chave — uma tarefa que é trivial para os LLMs contemporâneos

Quando dizemos que os modelos de linguagem "entendem", queremos dizer que eles podem processar e gerar texto de maneiras que parecem coerentes e relevantes, não que possuam consciência ou compreensão semelhantes às humanas.

Os LLMs são treinados em grandes quantidades de dados textuais. Esse treinamento em larga escala permite que os LLMs capturem informações contextuais mais profundas e sutilezas da linguagem humana.

Os LLMs melhoraram significativamente o desempenho em uma ampla gama de tarefas de PNL, incluindo tradução de texto, análise de sentimentos, respostas a perguntas e muito mais.

Os modelos de PNL anteriores eram tipicamente projetados para tarefas específicas, os LLMs demonstram uma proficiência mais ampla em uma ampla gama de tarefas de PNL.

A previsão da próxima palavra é sensata porque aproveita a natureza sequencial inerente da linguagem para treinar modelos na compreensão de contexto, estrutura e relacionamentos dentro do texto.

Os LLMs utilizam uma arquitetura chamada de transformador, que permite que eles prestem atenção seletiva a diferentes partes da entrada ao fazer previsões

image.png

O machine learning envolve o desenvolvimento de algoritmos que podem aprender e fazer previsões com base em dados sem serem explicitamente programados.

Em vez de escrever manualmente regras para identificar e-mails de spam, um algoritmo de machine learning é alimentado com exemplos de e-mails rotulados como spam e e-mails legítimos.

O deep learning é um subconjunto do machine learning que se concentra na utilização de redes neurais com três ou mais camadas.

Em contraste com o deep learning , o machine learning tradicional requer extração manual de características. Isso significa que especialistas humanos precisam identificar e selecionar as características mais relevantes para o modelo.

Retornando ao exemplo de classificação de spam, na aprendizagem de máquina tradicional, especialistas humanos podem extrair manualmente características do texto do e-mail, como a frequência de certas palavras-chave (por exemplo, “prêmio”, “ganhar”, “grátis”), o número de pontos de exclamação, o uso de palavras em maiúsculas ou a presença de links suspeitos. Este conjunto de dados, criado com base nessas características definidas por especialistas, seria então utilizado para treinar o modelo. Em contraste com a aprendizagem de máquina tradicional, a aprendizagem profunda não requer extração manual de características.


## 1.2 Aplicações de LLMs

Hoje, os LLMs são empregados para tradução automática, geração de textos novos (veja a figura 1.2), análise de sentimentos, resumo de textos e muitas outras tarefas. Os LLMs foram recentemente utilizados para criação de conteúdo, como escrever ficção, artigos e até mesmo código de computador.

Além disso, LLMs podem ser usados para recuperação eficaz de conhecimento a partir de vastos volumes de texto em áreas especializadas, como medicina ou direito. Isso inclui filtrar documentos, resumir trechos longos e responder a perguntas técnicas.

## 1.3 Estágios de construção e uso de LLMs

Por que devemos construir nossos próprios LLMs? Codificar um LLM do zero é um excelente exercício para entender sua mecânica e limitações. Além disso, nos fornece o conhecimento necessário para pré-treinamento ou ajuste fino de arquiteturas de LLM de código aberto existentes para nossos próprios conjuntos de dados ou tarefas específicas de domínio.

Pesquisas mostraram que, quando se trata de modelar desempenho, LLMs personalizados—aqueles adaptados para tarefas ou domínios específicos—podem superar LLMs de uso geral, como os fornecidos pelo ChatGPT, que são projetados para uma ampla gama de aplicações. Exemplos disso incluem BloombergGPT.

As empresas podem preferir não compartilhar dados sensíveis com provedores de LLM de terceiros, como a OpenAI, devido a preocupações de confidencialidade. Além disso, o desenvolvimento de LLMs personalizados menores permite a implantação direta em dispositivos dos clientes, como laptops e smartphones, além de conceder aos desenvolvedores total autonomia, permitindo que controlem atualizações e modificações no modelo conforme necessário.

O processo geral de criação de um LLM inclui pré-treinamento e ajuste fino. O "pré" em "pré-treinamento" refere-se à fase inicial onde um modelo como um LLM é treinado em um grande dataset diversificado para desenvolver uma ampla compreensão da linguagem. Este modelo pré-treinado serve então de base para poder ser refinado ainda mais por meio do ajuste fino, um processo onde o modelo é especificamente treinado em um dataset mais restrito que é mais específico para tarefas ou domínios particulares.

O pré-treinamento de um LLM envolve a previsão da próxima palavra em grandes conjuntos de dados de texto. Um LLM pré-treinado pode então ser ajustado usando um conjunto de dados rotulado menor.

## 1.4 Uma introdução a arquitetura transformer

Uma arquitetura de rede neural profunda introduzida no artigo de 2017 "Attention Is All You Need" (https://arxiv.org/abs/1706.03762).

Um componente chave dos transformers e LLMs é o self-attention mechanism (não mostrado), que permite ao modelo pesar a importância de diferentes palavras ou tokens em uma sequência em relação uns aos outros.

O aprendizado zero-shot refere-se à capacidade de generalizar para tarefas completamente desconhecidas sem exemplos específicos prévios. Por outro lado, o aprendizado few-shot envolve aprender a partir de um número mínimo de exemplos que o usuário fornece como entrada.
